```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib

matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = 28, 15

df = pd.read_csv('1031711.csv')
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%m%d')
df = df.set_index(['DATE'])
df = df.loc['2014-7-1':'2017-7-1',['PRCPT', 'TMAX', 'TMIN', 'YEAR']] #'2009-1-1':'2017-7-16'

cleaned = df.dropna(thresh=4)

tmax_pivoted = cleaned.pivot(columns='YEAR', values='TMAX')
axes_temp = tmax_pivoted.plot()
axes_temp.set_title('High Temps Per Day')
axes_temp.set_ylabel('Temperature Fahrenheit')

prcp_pivoted = cleaned.pivot(columns='YEAR', values='PRCPT')
axes_precip = prcp_pivoted.plot()
axes_precip.set_title('Precipitation Per Day')
axes_precip.set_ylabel('Precipitation Inches')
```
